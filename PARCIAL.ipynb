```
import pandas as pd

class Nodo:
    def __init__(self, Año, Localidad, Indicador, Periodo, siguiente=None, anterior=None):
        self.Año = Año
        self.Localidad = Localidad
        self.Indicador = Indicador
        self.Periodo = Periodo
        self.siguiente = siguiente
        self.anterior = anterior


class ListaEnlazada:
    def __init__(self):
        self.cabeza = None
        self.cola = None

    def agregar_nodo(self, nodo):
        if self.cabeza is None:
            self.cabeza = nodo
            self.cola = nodo
        else:
            self.cola.siguiente = nodo
            nodo.anterior = self.cola
            self.cola = nodo

    def mostrar_normal_y_almacenar_con_filtro(self):
        """
        Muestra los datos de la lista enlazada en orden normal,
        almacena datos dependientes en un diccionario y
        filtra por años 2017 y 2021.

        Returns:
            dict: Un diccionario donde las claves son los 'Indicador'
                  y los valores son listas de los 'Año' correspondientes.
        """

        datos_dependientes = {}
        nodo_actual = self.cabeza

        # First pass: Collect all nodes for 2017 and 2021
        nodes_2017 = []
        nodes_2021 = []
        while nodo_actual:
            if nodo_actual.Año == 2017:
                nodes_2017.append(nodo_actual)
            elif nodo_actual.Año == 2021:
                nodes_2021.append(nodo_actual)
            nodo_actual = nodo_actual.siguiente

        # Second pass: Display and store data for 2017 then 2021
        for nodo in nodes_2017 + nodes_2021:  # Concatenate lists to display in order
            print(
                f"Año: {nodo.Año}, Localidad: {nodo.Localidad}, Indicador: {nodo.Indicador}, Periodo: {nodo.Periodo}"
            )

            # Store dependent data (example: Año for each Indicador)
            indicador = nodo.Indicador
            if indicador in datos_dependientes:
                datos_dependientes[indicador].append(nodo.Año)
            else:
                datos_dependientes[indicador] = [nodo.Año]

        return datos_dependientes


# ... (rest of the code: reading CSV, creating linked list, calling the function remains the same)
# Leer archivo Excel
df_csv = pd.read_csv('Base de datos 2.csv', delimiter=';')

# Crear la lista enlazada
lista = ListaEnlazada()

# Iterar sobre las filas del DataFrame y crear nodos
for index, row in df_csv.iterrows():
    nuevo_nodo = Nodo(row[df_csv.columns[0]], row[df_csv.columns[1]], row[df_csv.columns[2]], row[df_csv.columns[3]])
    lista.agregar_nodo(nuevo_nodo)

# Llamar a la función para mostrar en orden normal, almacenar y filtrar datos
datos_almacenados = lista.mostrar_normal_y_almacenar_con_filtro()

# Imprimir los datos almacenados
print("\nDatos dependientes almacenados:")
print(datos_almacenados)
```
