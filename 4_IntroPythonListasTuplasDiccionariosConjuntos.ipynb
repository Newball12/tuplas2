## Listas (`list`)

Las listas son **estructuras de datos** que pueden almacenar cualquier otro tipo de dato, inclusive una lista
puede contener otra lista, además, la cantidad de elementos de ***una lista se puede modificar*** removiendo o
añadiendo elementos. Para definir una lista se utilizan los paréntesis cuadrados, dentro de estos se colocan todos
los elementos separados por comas:

```python
calificaciones = [10,9,8, 7.5, 9]
nombres = ["Ana","Juan","Sofía","Pablo","Tania"]
mezcla = [True, 10.5, "abc", [0,1,1]]
```

```python
lista = [1, ['a', 'e', 'i', 'o', 'u'], 8.9, 'hola']
```

Las listas son *iterables* y por tanto se puede acceder a sus elementos mediante indexación:

**RECORDATORIO**: Python usa Indexación en base-0

Recordemos que Python usa indexación en base-0. Esto quiere decir que el primer elemento en un objeto esta localizado en la posición 0. Esto es diferente de otros lenguajes como R y Matlab que indexan elementos dentro de objetos iniciando en 1.

```python
nombres[2]
```

## Tuplas (`tuple`)

Las tuplas son secuencias de elementos similares a las listas, la diferencia principal es que
las tuplas ***no pueden ser modificadas directamente***, es decir, una tupla no dispone de los métodos
como `append` o `insert` que modifican los elementos de una lista.

Para definir una tupla, los elementos se separan con comas y se encierran entre paréntesis.

```python
colores=("Azul","Verde","Rojo","Amarillo","Blanco","Negro","Gris")
```

Las tuplas al ser *iterables* pueden accederse mediante la notación de corchetes e índice.

### Diccionarios (`dict`)

Los diccionarios son estructuras que contienen una colección de elementos de la
forma `clave: valor` separados por comas y encerrados entre llaves.
Las claves deben ser objetos inmutables y los valores pueden ser de cualquier tipo.
Necesariamente las claves deben ser únicas en cada diccionario, no así
los valores.


*   Son secuencias no ordenadas de datos
*   No tiene un indice de python asociado
*   Son mutables por lo se pueden modificar sus valores
*   No se pueden seccionar
*   No se pueden concatenar ni realizar repitición

### Listas Doblemente Enlazadas

```python
# Clase para definir el nodo de la lista doblemente enlazada
class Nodo:
    def __init__(self, nombre, apellido, id, edad):
        self.nombre = nombre  # Contiene el nombre del animal
        self.apellido = apellido  # Contiene el apellido del animal
        self.id = id  # Contiene el id del animal
        self.edad = edad  # Contiene la edad del animal
        self.siguiente = None  # Apunta al siguiente nodo
        self.anterior = None  # Apunta al nodo anterior

# Clase para definir la lista doblemente enlazada
class ListaDobleEnlazada:
    def __init__(self):
        self.cabeza = None  # Inicializa la lista vacía

    # Método para insertar un nuevo nodo al final de la lista
    def insertar(self, nombre, apellido, id, edad):
        nuevo_nodo = Nodo(nombre, apellido, id, edad)
        if not self.cabeza:  # Si la lista está vacía
            self.cabeza = nuevo_nodo
        else:
            actual = self.cabeza
            while actual.siguiente:  # Recorre hasta el último nodo
                actual = actual.siguiente
            actual.siguiente = nuevo_nodo  # Establece el siguiente del último nodo
            nuevo_nodo.anterior = actual  # Apunta al nodo anterior

    # Método para mostrar los elementos de la lista de adelante hacia atrás
    def mostrar_adelante(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual:
            print(f"Informe: {actual.nombre}, {actual.apellido}, {actual.id}, {actual.edad}", end=" <-> ")
            actual = actual.siguiente
        print("None")  # Indica el final de la lista

    # Método para mostrar los elementos de la lista de atrás hacia adelante
    def mostrar_atras(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        # Ir al último nodo
        while actual.siguiente:
            actual = actual.siguiente
        # Recorrer hacia atrás
        while actual:
            print(f"Informe: {actual.nombre}, {actual.apellido}, {actual.id}, {actual.edad}", end=" <-> ")
            actual = actual.anterior
        print("None")  # Indica el inicio de la lista

# Crear una lista doblemente enlazada
lista_doble = ListaDobleEnlazada()

# Insertar algunos animales en la lista
lista_doble.insertar('alejandro','newball','123343','46')
lista_doble.insertar('andres','william','345456','67')
lista_doble.insertar('nafther','macnish','976567','30')

# Mostrar la lista de adelante hacia atrás
print("Lista de adelante hacia atrás:")
lista_doble.mostrar_adelante()

# Mostrar la lista de atrás hacia adelante
print("\nLista de atrás hacia adelante:")
lista_doble.mostrar_atras()
```

##Métodos para Ordenar Datos
Método Burbuja (Bubble Sort)

```python
lista = [64, 34, 25, 12, 22, 11, 90]
n = len(lista)

for i in range(n):
    for j in range(0, n-i-1):
        if lista[j] > lista[j+1]:
            lista[j], lista[j+1] = lista[j+1], lista[j]

print("Lista ordenada con Burbuja:", lista)
```

#Método de Selección (Selection Sort)
Descripción: Encuentra el elemento más pequeño en la lista y lo intercambia con el primer elemento no ordenado. Repite este proceso con los elementos restantes.

```python
lista = [64, 34, 25, 12, 22, 11, 90]

for i in range(len(lista)):
    min_idx = i
    for j in range(i+1, len(lista)):
        if lista[j] < lista[min_idx]:
            min_idx = j
    lista[i], lista[min_idx] = lista[min_idx], lista[i]

print("Lista ordenada con Selección:", lista)
```

#Método de Mergesort (Merge Sort)
Descripción: Divide la lista en sublistas hasta que cada sublista tenga un solo elemento, luego combina las sublistas ordenadas en una lista ordenada.
Complejidad:
𝑂
(
𝑛
log

𝑛
)
O(nlogn)
Uso: Es un algoritmo estable y eficiente, especialmente para listas grandes.

```python
def merge_sort(lista):
    if len(lista) > 1:
        mid = len(lista) // 2
        L = lista[:mid]
        R = lista[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                lista[k] = L[i]
                i += 1
            else:
                lista[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            lista[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            lista[k] = R[j]
            j += 1
            k += 1

lista = [64, 34, 25, 12, 22, 11, 90]
merge_sort(lista)
print("Lista ordenada con Mergesort:", lista)
```

#Método Quicksort (Quick Sort)
Descripción: Selecciona un "pivote" y particiona la lista en dos sublistas: una con elementos menores que el pivote y otra con elementos mayores. Luego, ordena las sublistas recursivamente.
Complejidad:
𝑂
(
𝑛

𝑛
)
O(nlogn) en promedio, pero
𝑂

2
 ) en el peor caso.
Uso: Es uno de los algoritmos de ordenación más rápidos en la práctica, aunque no es estable.

```python
def quick_sort(lista):
    if len(lista) <= 1:
        return lista
    else:
        pivot = lista[len(lista) // 2]
        left = [x for x in lista if x < pivot]
        middle = [x for x in lista if x == pivot]
        right = [x for x in lista if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)

lista = [64, 34, 25, 12, 22, 11, 90]
lista = quick_sort(lista)
print("Lista ordenada con Quicksort:", lista)
```

#Método Heapsort (Heap Sort)
Descripción: Convierte la lista en un heap (una estructura de datos similar a un árbol binario) y extrae repetidamente el elemento máximo o mínimo para formar la lista ordenada.
Complejidad:
𝑂
(
𝑛
log
⁡
𝑛
)
O(nlogn)
Uso: Es eficiente y tiene un rendimiento constante, pero no es estable.

```python
def heapify(lista, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and lista[i] < lista[l]:
        largest = l

    if r < n and lista[largest] < lista[r]:
        largest = r

    if largest != i:
        lista[i], lista[largest] = lista[largest], lista[i]
        heapify(lista, n, largest)

def heap_sort(lista):
    n = len(lista)

    for i in range(n // 2 - 1, -1, -1):
        heapify(lista, n, i)

    for i in range(n-1, 0, -1):
        lista[i], lista[0] = lista[0], lista[i]
        heapify(lista, i, 0)

lista = [64, 34, 25, 12, 22, 11, 90]
heap_sort(lista)
print("Lista ordenada con Heapsort:", lista)
```

#Ordenamiento por Conteo (Counting Sort)
Descripción: Es un algoritmo no comparativo que cuenta la frecuencia de cada elemento y luego calcula la posición de cada uno en la lista ordenada.
Complejidad:
𝑂
(
𝑛
+
𝑘
)
O(n+k) donde
𝑘
k es el rango de los números en la lista.
Uso: Es eficiente para listas con un rango pequeño de valores enteros.

```python
def counting_sort(lista):
    max_val = max(lista)
    m = max_val + 1
    count = [0] * m

    for a in lista:
        count[a] += 1
    i = 0
    for a in range(m):
        for c in range(count[a]):
            lista[i] = a
            i += 1

lista = [64, 34, 25, 12, 22, 11, 90]
counting_sort(lista)
print("Lista ordenada con Counting Sort:", lista)
```
