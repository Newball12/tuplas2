```
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Descargar recursos de NLTK si es necesario
nltk.download('punkt')
nltk.download('stopwords')

# Cargar los datos
df = pd.read_csv('tweets_dataset.csv')

# Función para limpiar y tokenizar un tweet
def limpiar_tokenizar(texto):
    # Tokenización
    tokens = word_tokenize(texto)

    # Convertir a minúsculas y eliminar stop words
    stop_words = set(stopwords.words('spanish'))  # Ajusta el idioma si es necesario
    tokens = [token.lower() for token in tokens if token not in stop_words]

    # Stemming (opcional)
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    return tokens

# Aplicar la función a la columna de tweets
df['tokens'] = df["texto"].apply(limpiar_tokenizar)

# Crear una nueva columna con los tokens unidos por espacios
df['texto'] = df['tokens'].apply(lambda x: ' '.join(x))

# Mostrar los primeros 5 resultados
print(df.head())

# Convertir el DataFrame a HTML (incluyendo la columna de tokens)
html_table = df.to_html()

# Mostrar la tabla en el navegador (si estás en un entorno Jupyter Notebook)
from IPython.display import HTML
display(HTML(html_table))

# Guardar la tabla en un archivo HTML
with open('tabla_tweets_limpios.html', 'w') as f:
    f.write(html_table)
```
